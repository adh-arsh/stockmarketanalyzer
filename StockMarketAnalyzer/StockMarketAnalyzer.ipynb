```python
import datetime
from datetime import date
from nsepy import get_history

import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
#block to get dates
today = str(date.today())
cyear = int(today[0:4])
cmonth = int(today[5:7])
cday = int(today[8:10])

#calculating the start date
a_date = datetime.date(cyear,cmonth,cday)
days = datetime.timedelta(300)
new_date = a_date - days
startday = str(new_date)

syear = int(startday[0:4])
smonth = int(startday[5:7])
sday = int(startday[8:10])
```

```python
#getting list of companies
table = pd.read_csv('companylist.csv',index_col=False)
listofcompanies = list(table['SYMBOL'])
```

```python
#to get the company's data
#finally to get the sorlisted companies

def getMA(table):
    numbers = table['Close']
    window_size = 44
    i = 0
    moving_averages = []
    
    while i < len(numbers) - window_size + 1:
        this_window = numbers[i : i + window_size]
        window_average = sum(this_window) / window_size
        moving_averages.append(window_average)
        i += 1
        len(moving_averages)
    return moving_averages

count = 0 #to count the number of iterations // companies scanned
today_values = pd.DataFrame()
for company in listofcompanies:
    
    
    #getting values of all companies
    bin = get_history(symbol=company, 
                      start=date(syear,smonth,sday), 
                      end=date(cyear,cmonth,cday))
    

    #getting moving average
    try:
        moving_averages = getMA(bin)
        row_start = bin.index[ int(bin['Close'].count()) - len(moving_averages) ]
        row_end = bin.index[ int(bin['Close'].count())-1 ]
        global bin44
        bin44 = bin.loc[row_start:row_end]
        bin44['Date'] = bin44.index
        bin44['MA44'] = moving_averages
        bin44['Mean'] = (bin['High'] + bin['Low'])/2
        
        row_start = bin44.index[ int(bin44['Close'].count()) - 1 ]
        row_end = bin44.index[ int(bin44['Close'].count())-1 ]
        list_dataframe = bin44.loc[row_start:row_end]
        
        today_values = today_values.append(list_dataframe)
        
    except:
        pass
    
    count +=1
    print(count,"of",len(listofcompanies))
    
    #debugging block delete later -----::::!!!!
    #if count == 1000:
       # break
    #debugging block delete later -----::::!!!!
    
sortlisted = today_values[(today_values['High'] >= today_values['MA44']) & (today_values['Low']<=today_values['MA44'])]
sortlisted
```

```python
#to show charts of shortlisted companies

for company in sortlisted['Symbol']:
    
    #getting values of all companies
    global bina
    bina = get_history(symbol=company, 
                      start=date(syear,smonth,sday), 
                      end=date(cyear,cmonth,cday))
    moving_averages = getMA(bina)
    row_start = bina.index[ int(bina['Close'].count()) - len(moving_averages) ]
    row_end = bina.index[ int(bina['Close'].count())-1 ]
    global bin44temp
    bin44temp = bina.loc[row_start:row_end]
    bin44temp['Date'] = bin44temp.index
    bin44temp['MA44'] = moving_averages
    bin44temp['Mean'] = (bina['High'] + bina['Low'])/2

    try:
        fig = go.Figure()
        fig.add_trace(go.Candlestick(x=bin44temp['Date'], 
                                     open=bin44temp['Open'],high=bin44temp['High'], 
                                     low=bin44temp['Low'],close=bin44temp['Close']))
        fig.add_trace(go.Scatter(x=bin44temp['Date'], y=bin44temp['MA44']))
        print('\t',company)
        fig.show()
    except:
        pass
```

```python
#to calculate and store order related details

sortlisted['BUYAT'] = (sortlisted['High']) + 1
sortlisted['TGT'] = (sortlisted['High']) + ((sortlisted['High'] - sortlisted['Low']) * 2)
sortlisted['STOPLOSS'] = sortlisted['Low']
sortlisted['PROFITPERC'] = ( ((sortlisted['High'] - sortlisted['Low']) * 2) / ((sortlisted['High']) + 1) ) * 100
```

```python
#to generate the order list excel sheet

orderlist = sortlisted[['Symbol','BUYAT','TGT','STOPLOSS','PROFITPERC']]
orderlist
orderlist.to_excel('orders.xlsx', sheet_name='a')
```
